```python
import FinanceDataReader as fdr
import pandas as pd

df1 = pd.read_csv('Bitcoin Historical Data.csv')

df1['Date'] = pd.to_datetime(df1['Date'])

df1 = df1.set_index(['Date'])
df1
```

```python
#df1 = df1.loc[:, ['Price']]
df1['Price'] = df1['Price'].str.replace(',', '').astype('float')
df1['Open'] = df1['Open'].str.replace(',', '').astype('float')
df1['High'] = df1['High'].str.replace(',', '').astype('float')
df1['Low'] = df1['Low'].str.replace(',', '').astype('float')
```

```python
df1
```

```python
for i in range(len(list(df1['Vol.']))):
    if list(df1['Vol.'])[i].endswith('K'):
        temp = list(df1['Vol.'])[i].replace('K', '')
        df1['Vol.'][i] = float(temp) * 1000
    else:
        temp2 = list(df1['Vol.'])[i].replace('M', '')
        df1['Vol.'][i] = float(temp2) * 1000000
        
df1['Vol.'] = df1['Vol.'].astype('int')
```

```python
df1.reset_index(inplace=True)
```

```python
from mpl_finance import candlestick2_ohlc
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
from pandas_datareader import data  
from datetime import datetime


fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
candlestick2_ohlc(ax, df1['Open'], df1['High'], 
                  df1['Low'], df1['Price'],
                  width=0.5, colorup='r', colordown='b')
```

```python

start_date = datetime(2021,1,1)
end_date = datetime(2021,3,28)

kospi_df = data.get_data_yahoo('^KS11', start_date, end_date) 
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
index = kospi_df.index.astype('str') # 캔들스틱 x축이 str로 들어감

kospi_df['MA3'] = kospi_df['Close'].rolling(3).mean()
kospi_df['MA5'] = kospi_df['Close'].rolling(5).mean()
kospi_df['MA10'] = kospi_df['Close'].rolling(10).mean()
kospi_df['MA20'] = kospi_df['Close'].rolling(20).mean()

# 이동평균선
ax.plot(index, kospi_df['MA3'], label='MA3', linewidth=0.7)
ax.plot(index, kospi_df['MA5'], label='MA5', linewidth=0.7)
ax.plot(index, kospi_df['MA10'], label='MA10', linewidth=0.7)

# X축 티커 숫자 20개로 제한
ax.xaxis.set_major_locator(ticker.MaxNLocator(20))

# 그래프 title과 축 이름 지정
ax.set_title('KOSPI INDEX', fontsize=30)
ax.set_xlabel('Date')

# 캔들차트 그리기
candlestick2_ohlc(ax, kospi_df['Open'], kospi_df['High'], 
                  kospi_df['Low'], kospi_df['Close'],
                  width=0.5, colorup='r', colordown='b')
ax.legend()
plt.grid()
plt.show()
```
